# Import del dataset

```python
import pandas as pd
import numpy as np
from sklearn import preprocessing 

import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

#importing dataset
australia = pd.read_csv("AUS.csv", header = 0)
australia.head()
```

# Rinominiamo i campi

```python
new_names = {'Date': 'date', 'Location': 'location','MinTemp': 'min_temp', 
            'MaxTemp': 'max_temp', 'Rainfall':'rainfall', 'Evaporation': 'evaporation', 'Sunshine': 'sunshine', 
            'WindGustDir': 'wind_gust_dir', 'WindGustSpeed': 'wind_gust_speed', 'WindDir9am': 'wind_dir_9am', 'WindDir3pm': 'wind_dir_3pm',
            'WindSpeed9am': 'wind_speed_9am', 'WindSpeed3pm': 'wind_speed_3pm',
            'Humidity9am':'humidity_9am', 'Humidity3pm':'humidity_3pm', 'Pressure9am':'pressure_9am', 'Pressure3pm':'pressure_3pm',
            'Cloud9am':'cloud_9am', 'Cloud3pm':'cloud_3pm', 'Temp9am':'temp_9am', 'Temp3pm':'temp_3pm',
            'RainToday': 'rain_today', 'RainTomorrow': 'rain_tomorrow' }
australia.rename(columns=new_names, inplace=True)
australia.head(8)
```

```python
#correzione del campo wind_dir e wind_speed: mettiamo 0 al posto di null ai campi dir se la velocitÃ  del vento Ã¨ 0 (non c'Ã¨ vento in quel momento)
australia['wind_dir_9am'] = australia[['wind_dir_9am','wind_speed_9am']].apply(lambda x: '0' if x['wind_speed_9am']==0 else x['wind_dir_9am'], axis=1)
australia['wind_dir_3pm'] = australia[['wind_dir_3pm','wind_speed_3pm']].apply(lambda x: '0' if x['wind_speed_3pm']==0 else x['wind_dir_3pm'], axis=1)

# rimuoviamo i gust perchÃ¨ sono presi dalle colonne sopra, tra il massimo della giornata
australia = australia.drop(['wind_gust_dir','wind_gust_speed',"sunshine","evaporation"], axis=1)

#togliamo tutte quelle cittÃ  che hanno almeno il 40% di null sul campo cloud3pm
location = australia.location.unique()
#print(location)
count = 0
for loc in location:
    #print(loc, len(australia[australia['location']==loc]))
    lung = len(australia[australia['location']==loc])
    if australia[australia['location']==loc].isnull().sum().cloud_3pm > lung * 0.4:
        australia = australia.drop(australia[australia['location'] == loc].index)
        count +=1
print(count)


#rimuoviamo la riga che ha meno di 16 campi non nulli
australia = australia.dropna(thresh = 16)
#imputation dei valori, andando a prendere il valore di quel campo alla riga precedente
australia = australia.ffill(axis=0)

#aggiungiamo la colonna state in funzione della cittÃ  

#stati:
sa = ['Adelaide', 'MountGambier', 'Nuriootpa', 'Woomera'] #australia meridionale
wa = ['Albany', 'Witchcliffe', 'PearceRAAF', 'PerthAirport', 'Perth', 'SalmonGums',
      'Walpole'] #australia occidentale
nsw = ['Albury', 'BadgerysCreek', 'Cobar', 'CoffsHarbour', 'Moree',
       'Newcastle', 'NorahHead', 'Penrith', 'Richmond', 'Sydney', 'SydneyAirport',
       'WaggaWagga', 'Williamtown', 'Wollongong'] #Nuovo Galles del Sud
qld = ['Brisbane', 'Cairns', 'GoldCoast', 'Townsville'] #Queensland
tas = ['Hobart', 'Launceston'] #Tasmania
vic = ['Ballarat', 'Bendigo', 'Sale', 'Melbourne', 'MelbourneAirport',
      'Mildura', 'Nhil', 'Portland', 'Watsonia', 'Dartmoor'] #Victoria
act = ['Camberra', 'Tuggeranong', 'MountGinini'] #territorio della capitale australiana
nt = [ 'AliceSprings', 'Darwin', 'Katherine', 'Uluru'] #Territorio del nord
nfk = ['NorfolkIsland'] #territorio esterno del norfolk
 
#gli ultimi quattro si potrebbero unire anche in un unico territorio del sud est australiano,
#a livello di clima dovrebbero essere molto simii potrebbe avere difficoltÃ  a classificarli

def State(a):
    if a in sa:
        return 'SA'
    elif a in wa:
        return 'WA'
    elif a in nsw:
        return 'NSW'
    elif a in qld:
        return 'QLD'
    elif a in tas:
        return 'TAS'
    elif a in vic:
        return 'VIC'
    elif a in act:
        return 'ACT'
    elif a in nt:
        return 'NT'
    elif a in nfk:
        return 'NFK'
    

#le = preprocessing.LabelEncoder()
#crime['category_id'] = le.fit_transform(australia['state'])

australia['state'] = australia.location.apply(lambda x: State(x))


#australia[australia['rainfall'] > 100].head(20)
```

```python
# rainfall nel tempo
australia['date'] = pd.to_datetime(australia['date'])
timedf = pd.DataFrame(data = australia[['rainfall','date']])
timedf = australia.set_index('date')

#A causa delle grandi quantitÃ  di dati, otterrÃ  una media mobile trimestrale con una piccola finestra.
rolling_mean = timedf['rainfall'].resample('Q').sum().rolling(window=2, center=True).mean()

# Ottenere una ponderazione esponenziale genera una linea di tendenza piÃ¹ fluida
exp_weighted_mean = timedf['rainfall'].resample('Q').sum().ewm(span=10).mean()

ax = rolling_mean.plot(lw=2.5, figsize=(14,7), color='orange')
exp_weighted_mean.plot(ax=ax, lw=2.5, color='navy')
#La media mobile esponenziale prende in considerazione tutti i dati della serie storica, con una ponderazione esponenzialmente decrescente. 
#La ponderazione dei primi dati della serie sarÃ  minore col passare del tempo fino a diventare infinitesimale, ma rientrerÃ  sempre nel calcolo. 
#Nel calcolo della EMA viene tenuta in considerazione tutta la storia
ax.set_ylabel('rainfall')
ax.set_title('Media mobile semplice rispetto alla media mobile esponenziale delle precipitazioni nel tempo')
plt.show()
```

```python
#RADAR CHART
# Creating dataframes for the value counts of wind direction.
australia_9am = pd.DataFrame(australia['wind_dir_9am'].value_counts())
australia_3pm = pd.DataFrame(australia['wind_dir_3pm'].value_counts())

# Setting same index to later merge.
australia_3pm['direction'] = australia_3pm.index
australia_9am['direction'] = australia_9am.index

# Reordering the directions.
reorderlist = ['N','NNW','NW','WNW',
               'W','WSW','SW','SSW',
               'S','SSE','SE','ESE',
               'E','ENE','NE','NNE']

australia_9am = australia_9am.reindex(reorderlist)

# Merging the two dataframes together
australia_dir = australia_9am.merge(australia_3pm)
australia_dir.head()

```

```python
# Creating a radar chart to visualise the difference in wind directions based on time.

# Creating the angles for the circle.
angles = np.linspace(0, 2*np.pi, 16, endpoint=False)
angles = np.concatenate((angles,[angles[0]]))

# Wind direction values reformatted to fit radar chart.
dir_am = australia_dir['wind_dir_9am'].values
dir_am = np.concatenate((dir_am,[dir_am[0]]))

dir_pm = australia_dir['wind_dir_3pm'].values
dir_pm = np.concatenate((dir_pm,[dir_pm[0]]))

# Creating labels for the chart.
#labels = australia_dir['direction'].values

fig = plt.figure(figsize=(10,8))
ax = fig.add_subplot(111, polar=True)

# Sets the grid location to a circle.
ax.set_thetagrids(angles * 180/np.pi) #AGGIUNGO COME SECONDO PARAMETRO LABELS, MA NON VA
ax.set_theta_zero_location("N")

# Plots the am directions.
ax.plot(angles, dir_am, linewidth=2)
ax.fill(angles, dir_am, alpha=0.25)

# Plots the pm directions.
ax.plot(angles, dir_pm, linewidth=2)
ax.fill(angles, dir_pm, alpha=0.25)

ax.set_title('Direzione del vento 9am vs 3pm', fontsize=16)


# Setting the minimum value so the radar is more readable.
ax.set_rmin(1500)
ax.grid(True)
```

```python
#SCELGO SU QUALI ATTRIBUTI FARE, CLOUD E UMIDITA 9 AM 
#in base alla velocitÃ  del vento e al rainfall predict sul tipo di pioggia
x = australia.iloc[:, [4,7]].values #PARTE DA 0 L'INDICE DELLE COLONNE 
print(x.shape)
```

```python
#K MEANS METHOD ---->ELBOW METHOD 
from sklearn.cluster import KMeans
wcss = []

for i in range(1,12):
    km = KMeans(n_clusters = i, init = 'k-means++', max_iter=300, n_init = 10, random_state = 0) #crea k mean con n cluster
    km.fit(x)
    wcss.append(km.inertia_) #INERTIA DICE QUANTO I CLUSTER RIESCONO AD ESSERE UNIFORMI TRA DI LORO (varianza)
    
plt.figure(figsize=(20,8))
plt.plot(range(1,12),wcss)
plt.title("Metodo Elbow", fontsize=20)
plt.xlabel("numero di cluster")
plt.ylabel("wcss")
#SI CREA DEI K MEANS CON UN K CRESCENTE, GRAFICA IL NUMERO DI CLUSTER CHE ABBIAMO CREATO , L'INERTIA DIMINUISCE
#SI PRENDE L'ULTIMO GOMITO COME NUM DI CLUSTER Ã¨ SUL 4-7(esempio se cambi attributi cambia) CHE Ã¨ IL NUMERO DI CLUSTER
```

```python
#YELLOW BRICK PER VEDERE LA CONFERMA DEL K
from yellowbrick.cluster import KElbowVisualizer
model = KMeans()
visualizer=KElbowVisualizer(model, k=(1,12))
plt.figure(figsize=(15,8))
visualizer.fit(x)
visualizer.show()
```

```python
#tra umiditÃ  e rainfall il gomito Ã¨ a 4!! mentre tra umiditÃ  e temperatura alle 3pm Ã¨ a 4
#CLUSTERING SULLE GIORNATE PIOVOSE

km = KMeans(n_clusters = 4, init = 'k-means++', max_iter = 300, n_init = 10, random_state = 0)
y_means = km.fit_predict(x)
#si crea scatterplot colorandolo se appartiene ad un certo cluster
plt.figure(figsize=(15,8))
plt.scatter(x[y_means == 0, 0], x[y_means == 0, 1], s = 5, c = 'pink', label = 'Pioviggine')#poco acqua e poco vento
plt.scatter(x[y_means == 1, 0], x[y_means == 1, 1], s = 5, c = 'yellow', label = 'Pioggia di Forte IntensitÃ ')#tanta pioggia vento variabile
plt.scatter(x[y_means == 2, 0], x[y_means == 2, 1], s = 5, c = 'cyan', label = 'Piccole precipitazioni') #poco vento e acqua
plt.scatter(x[y_means == 3, 0], x[y_means == 3, 1], s = 5, c = 'magenta', label = 'Vento teso ')#poca pioggia
#plt.scatter(x[y_means == 4, 0], x[y_means == 4, 1], s = 5, c = 'orange', label = 'Pioggia di  Media IntensitÃ ')
plt.scatter(km.cluster_centers_[:,0], km.cluster_centers_[:, 1], s = 50, c = 'blue' , label = 'centroid')

plt.title('K Means Clustering', fontsize = 20)
plt.xlabel('rainfull')
plt.ylabel('velocitÃ ')
plt.legend()
plt.show()
```

```python
#SILHOUETTE VISUALIZER CON YELLOW BRICKER PER VEDERE CHE I CLUSTER SIANO DISTANZIATI E OMOGENEI
from yellowbrick.cluster import SilhouetteVisualizer
model = KMeans(4, random_state=42)
visualizer=SilhouetteVisualizer(model,colors="yellowbrick")
plt.figure(figsize=(15,8))
visualizer.fit(x)
visualizer.show()
```

```python
#ALTRO ALGORITMO DI CLUSTERING HIERARCHIAL CLUSTERING
from sklearn.cluster import AgglomerativeClustering

hc = AgglomerativeClustering(n_clusters = 4, affinity = "euclidean", linkage = "ward")
y_hc = hc.fit_predict(x)

#si crea scatterplot colorandolo se appartiene ad un certo cluster
plt.figure(figsize=(15,8))
plt.scatter(x[y_hc == 0, 0], x[y_hc == 0, 1], s = 100, c = 'pink', label = 'miser')
plt.scatter(x[y_hc == 1, 0], x[y_hc == 1, 1], s = 100, c = 'yellow', label = 'general')
plt.scatter(x[y_hc == 2, 0], x[y_hc == 2, 1], s = 100, c = 'cyan', label = 'target')
plt.scatter(x[y_hc == 3, 0], x[y_hc == 3, 1], s = 100, c = 'magenta', label = 'spendthrift')
#plt.scatter(x[y_hc == 4, 0], x[y_hc == 4, 1], s = 100, c = 'orange', label = 'careful')
plt.scatter(km.cluster_centers_[:,0], km.cluster_centers_[:, 1], s = 50, c = 'blue' , label = 'centroid')

plt.title('Hierarchial Clustering', fontsize = 20)
plt.xlabel('Humidity 9am')
plt.ylabel('Cloud 9am')
plt.legend()
plt.show()
```
